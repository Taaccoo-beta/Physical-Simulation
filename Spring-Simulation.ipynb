## Spring Simulation

Spring function

- $\frac{d^{2}s}{dt^{2}}=-Ps(t)$


- $V = \frac{ds}{dt}$


- $s(t+\varDelta{t}) = s(t)+\frac{s(t)}{d(t)}*\varDelta{t}$

```python
import matplotlib.pyplot as plt
```

```python
dt = 0.05
p = -5.0
sp = 5.0

acc = [p*sp]
vel = [0.0]
s = [sp]
t = [0.0]

for i in range(1,1000):
    acc.append(s[-1]*p)
    vel.append(vel[-1]+acc[-1]*dt)
    s.append(s[-1]+vel[-1]*dt)
    t.append(dt*i)
plt.plot(t,s)
plt.show()
```
